```python
import torch 
import cv2
import os 
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
```

```python
test_path = 'F:/design/code/unet_42-master/test_predictions/'
label_path = test_path.replace('prediction','label')

test_path = glob.glob(os.path.join(test_path,'*'))
label_path = glob.glob(os.path.join(label_path,'*'))

num_patient = len(test_path)
```

```python
img_path = glob.glob(os.path.join(test_path[-2],'*.png'))
target_path = glob.glob(os.path.join(label_path[-2],'*.png'))

img1 = cv2.imread(img_path[32],0)
label1 = cv2.imread(target_path[32],0)

loss1 = abs(np.array(img1,dtype='int')-np.array(label1,dtype='int'))

img2 = cv2.imread(img_path[54],0)
label2 = cv2.imread(target_path[54],0)
loss2 = abs(np.array(img2,dtype='int')-np.array(label2,dtype='int'))

img3 = cv2.imread(img_path[64],0)
label3 = cv2.imread(target_path[64],0)
loss3 = abs(np.array(img3,dtype='int')-np.array(label3,dtype='int'))


fig, ax = plt.subplots(3, 3,figsize=(10,10))
ax[0][0].set_axis_off()
ax[0][0].imshow(img1,cmap='jet')
ax[0][0].set(title='Predicted Dose')
ax[0][1].set_axis_off()
ax[0][1].imshow(label1,cmap='jet')
ax[0][1].set(title='Planned Dose')
ax[0][2].set_axis_off()
ax[0][2].imshow(loss1,cmap='jet')
ax[0][2].set(title='Difference')

ax[1][0].set_axis_off()
ax[1][0].imshow(img2,cmap='jet')
ax[1][1].set_axis_off()
a = ax[1][1].imshow(label2,cmap='jet')
ax[1][2].set_axis_off()
ax[1][2].imshow(loss2,cmap='jet')

ax[2][0].set_axis_off()
ax[2][0].imshow(img3,cmap='jet')
ax[2][1].set_axis_off()
ax[2][1].imshow(label3,cmap='jet')
ax[2][2].set_axis_off()
ax[2][2].imshow(loss3,cmap='jet')
vmin = 0
vmax = 255
norm = matplotlib.colors.Normalize(vmin=vmin,vmax=vmax)
bar = fig.colorbar(a,ax = [ax[0][2],ax[1][2],ax[2][2]])
bar.set_label('Gy')


plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.86, 
                    top=0.9, 
                    wspace=0.0001, 
                    hspace=-0.15)

plt.savefig('bijiaotu.png')
```
